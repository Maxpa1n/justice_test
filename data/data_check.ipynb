```python
data0 = []
data1 = []
with open('0_train.json','r',encoding='utf8') as f:
    for i in f:
        data0.append(json.loads(i.replace('\n','')))
with open('1_train.json','r',encoding='utf8') as f:
    for i in f:
        data1.append(json.loads(i.replace('\n','')))
```

```python
lenth = []
for i,j in zip(data0,data1):
    lenth.append(len(i['statement']))
    lenth.append(len(j['statement']))
```

```python
data_all = []
for i,j in zip(data0,data1):
    data_all.append(i)
    data_all.append(j)
```

```python
len(data_all)
```

```python
import random
random.shuffle(data_all)
```

```python
# 1943
```

```python
for j in range(8):
    s = j*1943
    e = (j+1)*1943
    
    dev = data_all[s:e]
    train = data_all[:s]+ data_all[e:]
    with open('data_list/trian_index_{}.json'.format(j+1),'w',encoding='utf8') as f:
        for k in train:
            json.dump(k,f,ensure_ascii=False)
            f.write('\n')
    with open('data_list/dev_index_{}.json'.format(j+1),'w',encoding='utf8') as f:
        for k in dev:
            json.dump(k,f,ensure_ascii=False)
            f.write('\n')
```
